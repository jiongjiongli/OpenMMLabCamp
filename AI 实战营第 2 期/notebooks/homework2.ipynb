```python
from pathlib import Path

def list_file_suffixes(dir_path):
    dir_path = Path(dir_path)
    
    suffix_list = []
    
    for file_path in dir_path.rglob('*.*'):
        suffix_list.append(file_path.suffix)
        
    print(list(set(suffix_list)))

data_dir_path = 'data/cats_dogs_dataset'
list_file_suffixes(data_dir_path)
```
